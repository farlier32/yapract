```python
def games_plot(df, column, func, y='name', xl=None,yl=None):
    temp_df = df[df['year_of_release'] != 2222]
    temp_df2 = temp_df.groupby(column)[y]
    if func == 'count':
        calculate_df = temp_df2.count()
        plot = calculate_df.plot(kind='bar',
                                 y=y, figsize=(15,5),
                                 ec='black', color='pink',
                                xlabel=xl, ylabel=yl)
        
    elif func == 'sum':
        calculate_df = temp_df2.sum().sort_values()
        plot = calculate_df.plot(kind='barh',
                                 y=y,
                                 figsize=(15,10),
                                 ec='black', color='pink',
                                xlabel=xl, ylabel=yl)
```

## –°–æ—Å—Ç–∞–≤—å—Ç–µ –ø–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

### –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã (—Ç–æ–ø-5). –û–ø–∏—à–∏—Ç–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ –¥–æ–ª—è—Ö –ø—Ä–æ–¥–∞–∂.

```python
def regions_plot(df, groupby, region_sales, ax):
    pivot = df.groupby(groupby).agg({region_sales: 'sum'}).sort_values(by=region_sales, ascending=False)[:5]
    title_dict = {'na_sales': 'North America Sales', 'eu_sales': 'Europe Sales', 'jp_sales': 'Japan Sales'}
    color_dict = {'na_sales': 'blue', 'eu_sales': 'gray', 'jp_sales': 'forestgreen'}
    plot = pivot.plot(kind='bar', ec='black', title=title_dict[region_sales], ax=axes[ax], 
                   fontsize=18, color=color_dict[region_sales], rot=20)
    plot.legend(prop={'size':17})
    plot.set_xlabel('')
    plot.title.set_size(20)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25,6))

regions_plot(games, 'platform', 'na_sales', 0)
regions_plot(games, 'platform', 'eu_sales', 1)
regions_plot(games, 'platform', 'jp_sales', 2)
```
